# Paso 2: EDA (Exploratory Data Analysis)
## Proyecto: Detección de Fraude Financiero

En esta fase exploraremos el dataset limpio para encontrar patrones, distribuciones y relaciones entre las variables y el fraude.

### Metodología CRISP-ML: Comprensión de los Datos

### 2.1 Inspección Inicial
Cargamos `fraude_limpio.csv` y revisamos la información general.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargamos el dataset limpio generado en el paso anterior
df = pd.read_csv('fraude_limpio.csv')

# Información general del dataset
print("Estructura del dataset:")
df.info()

print("\nEstadísticas descriptivas:")
df.describe()
```

### 2.2 Visualización de Distribuciones
Analizamos la proporción de fraude en el dataset.

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='es_fraude', palette='viridis')
plt.title('Distribución de Fraude (0 = No, 1 = Sí)')
plt.xlabel('Es Fraude')
plt.ylabel('Cantidad')
plt.show()
```

### 2.3 Relación entre Producto y Fraude
¿Qué productos financieros tienen más reportes de fraude?

```python
plt.figure(figsize=(12, 8))
sns.countplot(data=df, y='producto', hue='es_fraude', order=df['producto'].value_counts().index[:10])
plt.title('Top 10 Productos con Mayor Reporte de Fraude')
plt.xlabel('Cantidad')
plt.ylabel('Producto')
plt.legend(title='Es Fraude')
plt.show()
```

### 2.4 Distribución Geográfica (Departamentos)
Identificamos las zonas con mayor incidencia de quejas.

```python
plt.figure(figsize=(12, 6))
df['departamento'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Quejas por Departamento')
plt.xticks(rotation=45)
plt.show()
```
