# Paso 1: ETL (Extract, Transform, Load)
## Proyecto: Detección de Fraude Financiero

En esta fase realizaremos la extracción de datos brutos, su transformación para limpieza y enriquecimiento, y finalmente la carga de un archivo limpio listo para el análisis.

### Metodología CRISP-ML: Preparación de Datos

### 1.1 Extracción (Extract)
Cargamos el dataset original `fraude.csv` utilizando la librería pandas.

```python
import pandas as pd
import numpy as np

# Cargamos el dataset original
ruta_archivo = 'fraude.csv'
df_original = pd.read_csv(ruta_archivo)

# Visualizamos las primeras filas para entender la estructura
df_original.head()
```

### 1.2 Transformación (Transform)
Realizaremos las siguientes tareas:
1. Limpieza de nombres de columnas.
2. Manejo de valores nulos.
3. Creación de la variable objetivo `es_fraude` basada en el motivo de la queja.
4. Filtrado de columnas no necesarias para el modelo predictivo.

```python
# 1. Renombrar columnas para estandarizar (quitar tildes y espacios si hubiera)
df_transformed = df_original.copy()

# 2. Manejo de nulos: Eliminamos filas donde el departamento o municipio sea nulo
antes = len(df_transformed)
df_transformed = df_transformed.dropna(subset=['departamento', 'municipio'])
print(f"Filas eliminadas por nulos: {antes - len(df_transformed)}")

# 3. Definición de Fraude: 
# Consideramos fraude los motivos de 'SUPLANTACIÓN O FRAUDE' y 'Transacción no reconocida'
motivos_criticos = ['SUPLANTACIÓN O FRAUDE', 'Transacción no reconocida']
df_transformed['es_fraude'] = df_transformed['motivo'].apply(lambda x: 1 if x in motivos_criticos else 0)

# 4. Selección de características relevantes
columnas_a_mantener = [
    'tipo_entidad', 
    'codigo_entidad', 
    'instancia_recepcion', 
    'producto', 
    'departamento', 
    'cantidad_quejas_recibidas', 
    'es_fraude'
]
df_final = df_transformed[columnas_a_mantener]

print("Transformación completada con éxito.")
df_final.head()
```

### 1.3 Carga (Load)
Guardamos el dataset limpio en un nuevo archivo CSV llamado `fraude_limpio.csv`.

```python
# Guardamos el resultado para ser usado en los siguientes cuadernos
df_final.to_csv('fraude_limpio.csv', index=False, encoding='utf-8')
print("Archivo 'fraude_limpio.csv' generado correctamente.")
```
